⏬  Je fais une Nearby recherche dans la ville de Meaux afin de récupérer des infos(dont le place_id) des établissement de type hôpitaux. Pour ce faire, j'effectue une requête à l'API GooglePlaces

```python
import requests
import googlemaps


YOUR_API_KEY="ME_DEMANDER_LA_CLE_EN_PRIVE"


url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=48.956202%2C2.888466&radius=4000&type=hospital&key="+YOUR_API_KEY

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
```

⏬La **réponse** à notre requête est format **JSON** comme demandé, elle **liste les établissements trouvé avec leurs différents informations** *(rating, user_ratings (ca nous intéresse),**place_id**(qui va nous permettre par la suite de récupérer les avis sur les établissements),location...)*

```python
print(response.text)
```

⏬ On récupère tout les places id de notre réponse ci-dessus que l'on stocke dans une liste

```python
#-------------------------------------------------Recuperer la liste des place_id-------------------------------------------------
import json

# Extraire les données JSON de la réponse
data = json.loads(response.text)

# Récupérer les place_id
place_ids = []
for result in data['results']:
    place_ids.append(result['place_id'])

# Afficher les place_ids
print(place_ids)
```

⏬Connaissant les places ids des etablissements qui nous intéressent, nous pouvons effectuer la requête nous permettant de récupérer les avis des citoyens. 

```python
#------------------------------------------------Recuperer les avis de chaque place_id ------------------------------------------
response_list=[]
reviews_list=[]

for place_id in place_ids:
  url = "https://maps.googleapis.com/maps/api/place/details/json?place_id="+place_id+"&language=fr&fields=name%2Crating%2Cformatted_phone_number%2Creviews&newest&key="+YOUR_API_KEY

  payload={}
  headers = {}

  response = requests.request("GET", url, headers=headers, data=payload)

  response_list.append(response)
```

⏬Nous avons à notre disposition une liste contenant les infos chaque établissements suivi de ces avis (si il en a) les plus récents.
